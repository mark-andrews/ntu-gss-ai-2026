```python
import numpy as np
x = np.array([1.5, 2.0, -1.0])
w = np.array([0.5, -0.3, 0.8])
b = 0.1
```

```python
# weighted sum of input + bias
np.sum(x * w) + b
```

```python
np.dot(x, w) + b
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def tanh(x):
  return np.tanh(x)
```

```python
sigmoid(2)
```

```python
sigmoid(-2)
```

```python
sigmoid(20)
```

```python
sigmoid(np.dot(w, x) + b)
```

```python
tanh(np.dot(x, w) + b)
```

```python
def neuron(x, w, b, activation):
  return activation(np.dot(x, w) + b)
```

```python
neuron(x, w, b, sigmoid)
```

```python
neuron(x, w, b, tanh)
```

```python
neuron(x, w, b, sigmoid)
```

```python
I = 10000 # input dimension
H = 50  # hidden dimension
O = 10  # dimension of output
N = 500 # number of images
```

```python
# define random weights
np.random.seed(101)
W1 = np.random.randn(I, H)
W2 = np.random.randn(H, O)
B1 = np.random.randn(H)
B2 = np.random.randn(O)

# define random input data
X = np.random.randn(N, I)
```

```python
X.shape
```

```python
hidden = sigmoid(X @ W1 + B1) # hidden layer representation
output = sigmoid(hidden @ W2 + B2) # output layer representation
```
