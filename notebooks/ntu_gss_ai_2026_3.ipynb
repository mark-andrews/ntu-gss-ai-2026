```python
from torchvision import datasets, transforms
import torch
import matplotlib.pyplot as plt
```

```python
train_data = datasets.MNIST(root='data',
                            download=True,
                            train=True,
                            transform=transforms.ToTensor())
```

```python
img, label = train_data[0] # first image and its label
```

```python
plt.imshow(img[0], cmap='gray', interpolation = 'none')
```

```python
plt.imshow(train_data[10][0][0],cmap='gray')
```

```python
def plot_digit(i, dataset = train_data):
  img, label = dataset[i]
  plt.imshow(img[0], cmap = 'gray')
  plt.title(label)
```

```python
plot_digit(2000)
```
